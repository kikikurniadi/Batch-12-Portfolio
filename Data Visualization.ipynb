# 1. Count of one categorical feature

```python
SaleCondition_count = df.groupby(['Neighborhood'],as_index=False).size()
```

```python
sns.barplot(x=SaleCondition_count['Neighborhood'],
            y=SaleCondition_count['size'])
```

```python
# Construct a Dataset
SaleCondition_count = df.groupby(['Neighborhood'],as_index=False).size()
SaleCondition_count = SaleCondition_count.sort_values(['size'],ascending=False)
```

```python
sns.barplot(x=SaleCondition_count['Neighborhood'],
            y=SaleCondition_count['size'])
plt.show()
```

```python
sns.barplot(x=SaleCondition_count['size'],
            y=SaleCondition_count['Neighborhood'])
plt.show()
```

# 3. Categorical-continuous

```python
sns.boxplot(x=df['SaleCondition'],
            y=df['YrSold'])
```

```python
# Construct data
def is_auto(Sale):
    if 'Normal' in Sale:
        return 'Normal'
    else:
        return 'Non Normal'
```

```python
df['Sale'] = df['SaleCondition'].apply(is_auto)
df['Sale'].value_counts()
```

```python
sns.boxplot(x=df['Sale'],
            y=df['YrSold']
           )
```
